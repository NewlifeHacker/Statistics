# Решающие деревья

```python
new = data.drop(['Name','SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis=1)

new = new.dropna(axis = 0)
new = new.dropna(axis = 1)
new['Sex'] = new['Sex'].replace({'female': 1} , regex = True)
new['Sex'] = new['Sex'].replace({'male': 0} , regex = True)
new
```

## Наиболее важные признаки получилось Fare and Sex

```python
from sklearn.tree import DecisionTreeClassifier

X_train = new[['Pclass', 'Fare', 'Age', 'Sex']]
Y_train = new['Survived']


clf = DecisionTreeClassifier(random_state=241)
clf.fit(X_train,Y_train)
importance = clf.feature_importances_
importance
```
