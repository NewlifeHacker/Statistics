```python
B_sort = sorted(B)
B_sort.sort()
Por_sort = Por.copy()
Por_sort.sort()
print("Вариционный ряд проницаемости - ", sorted(B))

print("Вариционный ряд пористости - ", sorted(Por))
```

## ЭФР

```python
ecdf = ECDF(B_sort)
ecdf2 = ECDF(Por_sort)
x1 = [0,min(B_sort)]
y1 = [0,0]
x2 = [0,min(Por_sort)]
y2 = [0,0]

fig1 = plt.figure(figsize=(16, 5))
efr = fig1.add_subplot(1, 2, 1)
efr2 = fig1.add_subplot(1,2,2)

efr.step(ecdf.x, ecdf.y, x1,y1 ,where='post', c= 'b')
efr2.step(ecdf2.x, ecdf2.y, x2,y2 ,where='post', c= 'b')

efr.set_title('Э.Ф.Р. проницаемости')
efr2.set_title('Э.Ф.Р. пористости')

plt.show()
```

### Поиск Квантилей и медианы




```python
quartilies = statistics.quantiles(B_sort, n=4)
print('Проницаемость-')
print("Первый квантиль- ", round(quartilies[0],3))
print("Медиана- ", round(quartilies[1],3))
print("Третий квантиль- ",round(quartilies[2],3))
print()
print('Пористость-')
print()
quartilies2 = statistics.quantiles(Por_sort, n=4)

print("Первый квантиль- ", round(quartilies2[0],3))
print("Медиана- ", round(quartilies2[1],3))
print("Третий квантиль- ",round(quartilies2[2],3))

```

## Вычисление выборочных моментов , дисперсия, стандартное отклонение

```python
print('Проницаемость ')
print()
print('Математическое ожидаение -', round(np.average(B_sort),3))
print('Дисперсия -', round(np.var(B_sort),3))
print('Стандартное (среднеквадратичное) отклонение - ', round(np.var(B_sort)**(1/2),3) )
print('Коэфициент вариации-  ', round(np.var(B_sort)**(1/2)/np.average(B_sort),3) )
print()
print('Пористось')
print()
print('Математическое ожидаение -', round(np.average(Por_sort),3))
print('Дисперсия -', round(np.var(Por_sort),5))
print('Стандартное (среднеквадратичное) отклонение - ', round(np.var(Por_sort)**(1/2),3) )
print('Коэфициент вариации-  ', round(np.var(Por_sort)**(1/2)/np.average(Por_sort),3 ))
```
